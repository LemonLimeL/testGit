# 데이터수집
#### 클로링 방법을 이용하여 수집 함
#### py 로 작성하여 됨.

````python
import urllib.request
from bs4 import BeautifulSoup
from itertools import count 
import pandas as pd
import ssl
import datetime
````

````python
def get_request_url(url, enc='utf-8'):
    context = ssl._create_unverified_context()
    req = urllib.request.Request(url)
    try:
        respose = urllib.request.urlopen(req, context=context)
        if respose.getcode() == 200:
            print("[%s] Url Request Success"% datetime.datetime.now())
            try:
                rcv = respose.read()
                ret = rcv.decode(enc)
            except UnicodeDecodeError:
                ret = rcv.decode(enc, 'replace')
            return ret
    except Exception as e:
        print(e)
        print("[%s] Error for URL " % datetime.datetime.now())
````

````python
def getBhcAdress(sido, result):
    for page in count():
        context = ssl._create_unverified_context()
        url = "http://www.bhc.co.kr/location/search.asp?branch_name=&gu=&si=%s&page=%s" \
        %(urllib.parse.quote(sido), str(page+1))
        rcv_data = get_request_url(url)
        soupData = BeautifulSoup(rcv_data, "html.parser")
        store_table = soupData.find('table', {'class':'register01'})    
        tbody = store_table.find('tbody')
        bEnd = True
        for store in tbody.findAll('tr'):
            bEnd = False;
            tr_tag = list(store.strings)
            name = tr_tag[6]
            phone = tr_tag[14].strip()
            adress = tr_tag[11]
            sido_gu = adress.split()[:2]
            result.append([name]+ sido_gu+[adress]+[phone])
        if(bEnd == True):
            return;
````

````python
sido_list = ['부산광역시']
result = [];

print("bhc 주소 크롤링 시작")
for sido in sido_list:
    getBhcAdress(sido, result)
print(result)
perincana_table = pd.DataFrame(result, columns=('store', 'sido', 'gungu', 'address', 'phone'))
perincana_table.head()
perincana_table.to_csv('bhc.csv', encoding='cp949',index =True)                
print("bhc 주소 크롤링 종료")

````
